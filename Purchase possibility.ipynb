```python
from sklearn import tree
from sklearn.datasets import load_iris
```

```python
X = [[1, 3, 2, 1],[1, 3, 2, 2],[2, 3, 2, 1],[3, 2, 2, 1],[3, 1, 1, 1],
     [3, 1, 1, 2],[2, 1, 1, 2],[1, 2, 2, 1],[1, 1, 1, 1],[3, 2, 1, 1],
     [1, 2, 1, 2],[2, 2, 1, 2],[2, 3, 1, 1]]
Y = [2,2,1,1,1,2,1,2,1,1,1,1,1]
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X,Y)
```

```python
tree.plot_tree(clf)
```

```python
clf.predict([[3, 2, 2, 2]])
```
